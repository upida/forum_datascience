```python
import pandas as pd
import seaborn as sns;

df = pd.read_csv("Health_Sleep_Statistics.csv")

df['Bedtime'] = pd.to_datetime(df['Bedtime'], format='%H:%M').dt.hour + pd.to_datetime(df['Bedtime'], format='%H:%M').dt.minute / 60
df['Wake-up Time'] = pd.to_datetime(df['Wake-up Time'], format='%H:%M').dt.hour + pd.to_datetime(df['Wake-up Time'], format='%H:%M').dt.minute / 60

# Converting gender to numeric
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == 'm' else 0)

# Converting sleep disorders to numeric
df['Sleep Disorders'] = df['Sleep Disorders'].apply(lambda x: 1 if x == 'yes' else 0)

# Converting medication usage to numeric
df['Medication Usage'] = df['Medication Usage'].apply(lambda x: 1 if x == 'yes' else 0)

correlation_sleep_bedtime = df[['Sleep Quality', 'Bedtime', 'Wake-up Time', 'Age', 'Gender', 'Sleep Disorders', 'Medication Usage']].corr()

sns.heatmap(correlation_sleep_bedtime, annot=True, cmap='YlGnBu')
```
