```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Health_Sleep_Statistics.csv")

data = data.sort_values(by=["Sleep Quality", "Bedtime"], ascending=[False, True])

plt.bar(data["Sleep Quality"], data["Bedtime"], color='skyblue')

plt.xlabel("Sleep Quality")
plt.ylabel("Bedtime")
plt.title("Sleep Quality vs Bedtime")

plt.show()
```
